```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Watches_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
```

```python
from pyspark.sql.functions import to_date
# Read in the Review dataset as a DataFrame
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
# Create the vine_table. DataFrame
vine_df = df.select(['review_id','star_rating','helpful_votes','total_votes','vine','verified_purchase'])
vine_df.show()
```

```python
vine_df_helpful=vine_df.filter(vine_df["total_votes"]>=20)
vine_df_helpful.show()
```

```python
vine_df_helpful_R=vine_df_helpful.filter(vine_df_helpful["helpful_votes"]/vine_df_helpful["total_votes"]>=0.5)
vine_df_helpful_R.show()
```

```python
vine_paid_df=vine_df_helpful_R.filter(vine_df_helpful_R["vine"]=="Y")
vine_paid_df.show()
```

```python
vine_unpaid_df=vine_df_helpful_R.filter(vine_df_helpful_R["vine"]=="N")
vine_unpaid_df.show()
```

```python
from pyspark.sql.functions import count, col, when, round
#vine_paid_df.select((count($'total_votes').as('Vote_Count'))).show()
vine_paid_df.select(count(vine_paid_df['star_rating']).alias("Total_Rating_Count"),
                    count(when(col('star_rating') == 5, True)).alias("Total_5_Star_Ratings"),
                    round(count(when(col('star_rating') == 5, True))/count(vine_paid_df['star_rating'])*100,2).alias("Percent_5_Star")).show()

vine_unpaid_df.select(count(vine_unpaid_df['star_rating']).alias("Total_Rating_Count"),
                    count(when(col('star_rating') == 5, True)).alias("Total_5_Star_Ratings"),
                    round(count(when(col('star_rating') == 5, True))/count(vine_paid_df['star_rating'])*100,2).alias("Percent_5_Star")).show()
```
